```python
def booking_trip(transport, destination):#keyword
  print(f"booking a {transport} trip to {destination}!")

booking_trip(transport="train", destination="kashmir")
```

```python
def booking_trip(transport, destination):#position
  print(f"booking a {transport} trip to {destination}!")

booking_trip("flight", "london")
```

```python
def booking_trip(transport, destination="london"):#default
  print(f"booking a {transport} trip to {destination}!")

booking_trip("flight")
```

```python
def booking_trip(transport="flight", destination="london"):#default
  print(f"booking a {transport} trip to {destination}!")

booking_trip("flight")
```

```python
def booking_trip(*transport, destination="london"):#arbitary
  print(f"booking a {transport} trip to {destination}!")

booking_trip("train",destination="london")
```
